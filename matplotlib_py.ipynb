BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["CLAD", "IT", "C", "ENGLISH","SINGLE","SEVERAL","GRAPHICS","PHYSICS","PRACTICALS","VDC"])
y = np.array([89, 97, 83, 85,89,82,97,72,94,76])

plt.bar(x,y)
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([89, 97, 83, 85,89,82,97,72,94,76])
mylabels = ["CLAD", "IT", "C", "ENGLISH","SINGLE","SEVERAL","GRAPHICS","PHYSICS","PRACTICALS","VDC"]

plt.pie(y, labels = mylabels)
plt.show() 
```

# exploding the pie chart graph

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["bmw", "mercedes", "maseratti", "rolls royce"]
yexplode = [0.5, 0.4, 0.3, 0.7]

plt.pie(y, labels = mylabels, explode = yexplode)
plt.show() 
```

Givnig colours to a pie chart

```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as kb
y = np.array([25,56,78,89])
mycars=["skoda","indigo","zen","alcazar"]
mycolors=["black","#ff6347","red","yellow"]
plt.pie(y,labels=mycars,colors=mycolors)
plt.show()
```
